### Load API Reponse (Postman)

```python
with open('nlp_response/1.txt', 'r') as f:
    response = f.read().replace('\n', '')
    
response = response[:-1]
response = '[' + response + ']'
nlp_list = ast.literal_eval(response)
```

```python
df1 = pd.DataFrame(nlp_list)
print(df1.shape)
df1.head()
```

```python
# Adding text classification to the comments dataframe
df_threat = pd.concat([df1, df0.head(20000)], axis=1)
```

```python
# rounding values to the nearest integer
df_threat[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']] = \
df_threat[['defense_AH', 'support_AH', 'offense_AH', 'defense_against_AH']].round()
```

```python
# change the type of the rounded values to boolean
df_threat['defense_AH'] = df_threat['defense_AH'].astype('bool')
df_threat['support_AH'] = df_threat['support_AH'].astype('bool')
df_threat['offense_AH'] = df_threat['offense_AH'].astype('bool')
df_threat['defense_against_AH'] = df_threat['defense_against_AH'].astype('bool')
```
